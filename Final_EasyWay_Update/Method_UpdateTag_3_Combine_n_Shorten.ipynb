```python
###########################################################################
#[1]　合併結果


def CombineAnCTAG(T4A,T4C):
    Combine = {}#　合併的大字典

    ###　　把評論的標籤結果給Combine
    for i in T4A:
        Combine[i]=T4A[i]
    ###　　
    c1 = 0#　只有評論有
    c2 = 0#　遊記有、評論也有的景點數量

    ###　　把評論的標籤結果給Combine
    for i in T4C:

    #######　　如果此時景點還不在Combine裡
        if not i in Combine:
            c1 += 1
            Combine[i]=T4C[i]#　就把內容加進去

    #######　　如果已經在Combine裡(代表此景點遊記.評論都有),就把內容補充進去    
        else:
            c2 += 1

    ###　　將T4C的內容放進Combine裡面        
            for a in T4C[i]:#　讀取T4C
                if a == u'概　念':#　如果內容是[概　念]
                    for b in T4C[i][a]:#　讀取各概念(ex. 古老.懷舊...等)
                        if not b in Combine[i][a]:#　如果這個概念還不在Combine裡
                            Combine[i][a][b]=T4C[i][a][b]#　就把概念的內容全部給Combine
                        else:#　如果這個概念 "已經在" Combine裡
                            for c in T4C[i][a][b]:#　讀取此概念裡面的詞(ex. 佛教.寺廟.宗教...等)
                                if not c in Combine[i][a][b]:#　如果概念裡面的詞還不在Combine的概念裡
                                    Combine[i][a][b][c]=T4C[i][a][b][c]#　就把詞放到Combine的該概念裡
                else:#　如果內容是[文字雲]
                    for b in T4C[i][a]:#　讀取文字雲內容
                        if not b in Combine[i][a]:#　如果這個詞語還不在Combine的文字雲裡
                            Combine[i][a].append(b)#　就把它加進去

    print '只有評論才有的景點數量:',c1,'\n',\
          '評論和遊記都有的景點數:',c2
    print '所有景點:',len(Combine)
      
    c1=0
    c2=0
    c3=0
    c4=0
    for i in Combine:
        if len(Combine[i][u'概　念'])==0 and len(Combine[i][u'文字雲'])==0:
            c1+=1
        if len(Combine[i][u'文字雲'])==0:
            c2+=1
        if len(Combine[i][u'概　念'])==0:
            c3+=1
        else:
            c4+=1

    print '\n1.甚麼都沒有的景點數量:',c1,'個\n',\
            '2.沒有「文字雲」的景點:',c2,'個\n',\
            '3.沒有「概　念」的景點:',c3,'個\n',\
            '4.概念文字雲都有的景點:',c4,'個\n',\
            '5.有「概　念」的景點數:',len(Combine)-c3,'個\n'

    if c1==c2:
        print '沒有文字雲，幾乎就等於甚麼都沒有！'
    if c4==(len(Combine)-c3):
        print '有概念的，幾乎就等於概念和文字雲都有！'

    return(Combine)


###########################################################################
#[2]　合併標籤


def ShortenTags(Combine,DicNew):
    CombineConcept={}#　縮減標籤後的字典
    for i in Combine:
        CombineConcept[i]={}
        ###　合併　############
        for a in Combine[i]:
            if a == u'概　念':#　概念
                CombineConcept[i][a]={}#　命景點的概念為一字典
                for b in Combine[i][a]:#　讀取各概念
                    for z in DicNew:#　讀取新的概念字典
                        if b in DicNew[z]:#　如果此概念在新概念裡
                            if not z in CombineConcept[i][a]:#　如果此景點還沒有這個新概念
                                CombineConcept[i][a][z]=Combine[i][a][b]#　就把東西複製過去
                            else:#　如果已經有新概念了
                                for c in Combine[i][a][b]:#　就把舊概念中的詞語
                                        CombineConcept[i][a][z][c]=Combine[i][a][b][c]#　放進新概念內
            else:#　文字雲
                CombineConcept[i][a]=Combine[i][a]  

    if len(Combine)==len(CombineConcept):
        print 'Trans To ShortenConcept Successfully! AttrNum is',len(CombineConcept)

    return CombineConcept


###########################################################################
#[3]　檢視結果


def WatchTagResultForDic(aa):#　字典版
    ##########################################################
    DicCT = {}#　存放概念的字典(KEY:景點　VALUE:標籤(陣列))
    ConDic = {}#　計算標籤量的字典(KEY:標籤　VALUE:數量)
    AttrNoTags = {}#　存放沒標籤的景點
    for i in aa:#　讀取景點
        for u in aa[i]:#讀取景點的內容
            
#           ###################################################
            if u == u'概　念':#　注意！因為檔案都是從JSON讀進來再併的!
                             #　所以格式都是Unicode！要特別注意！
#           ###################################################

                DicCT[i] = []
                if len(aa[i][u])==0:#　如果沒有概念的
                    #print i,aa[i][u]
                    AttrNoTags[i] = aa[i][u]#　加進AttrNoTags裡
                else:#　有概念的
                    for z in aa[i][u]:#　　把概念讀出來
                        DicCT[i].append(z)#　加進字典的VALUE裡
                        if not z in ConDic:#　如果這個概念第一次被讀到
                            ConDic[z] = 1#　加進字典裡
                        else:
                            ConDic[z]+= 1#　字典的計數加一

    print '景點數量:',len(aa)
    print '標籤種類:',len(ConDic)

    c = 0
    ###　　　標籤數　　　###
    for i in ConDic:
        print i,ConDic[i]
        c+=ConDic[i]

    print   '標籤總數:',c
    print   '有　標籤的景點數:',len(aa)-len(AttrNoTags)
    print   '總景點平均標籤數:',float(c)/float(len(aa))#　標籤數/全部景點
    print   '淨景點平均標籤數:',float(c)/float(len(aa)-len(AttrNoTags))#　標籤數/有標籤的景點

    print   '沒有標籤的景點數:',len(AttrNoTags),'\n',\
            '成功標籤比率:',1.0-(float(len(AttrNoTags))/float(len(aa)))

    print '========檢視沒有標籤的景點========'
    for i in AttrNoTags:
        print i

    ###　　把所有景點的標籤寫進一陣列裡（用來做關聯規則）　　###
    F = []
    for i in DicCT:
        F.append(DicCT[i])
    #return F


###########################################################################
#[4]　輸出文字雲


#A 文字雲字典
def DumpWC(Result,WordCloudFile):
    import json
    x = 0
    y = 0
    WordCloud = {}
    for a in Result:#　讀取結果的各個景點
        for b in Result[a]:#　讀取各個景點的Value
            if b == u'文字雲':
                for c in Result[a][b]:#　讀取文字雲的內容(陣列)
                    if not c in WordCloud:#　如果是第一次讀到
                        WordCloud[c] = 1#　則加進字典, 計算次數為一
                    else:#　如果已經讀過了
                        WordCloud[c] += 1#　則字典該詞語次數加一
                x += 1#　計算景點數
                y += len(Result[a][b])#　計算文字雲的字彙量

    z = 0
    for e in WordCloud:
        z += WordCloud[e]#　計算文字雲的字彙量

    if z==y:
        j_str = json.dumps(WordCloud)#　丟進一JSON格式的變數
        with open(WordCloudFile,'w') as f:#　寫進Json檔裡
            f.write(j_str)
        print 'Attr :',x,'Words :',len(WordCloud),'TotalWords :',z
        print 'Trans To WordCloud Dict Successfully!'
    
    
#B 概念文字雲字典
def DumpCC(CombineConcept,ConceptCloudFile):
    ConceptCloud={}
    b1 = 0
    b2 = 0
    for i in CombineConcept:
        for a in CombineConcept[i]:
            if a == u'概　念':
                for b in CombineConcept[i][a]:
                    b1 += 1
                    if b not in ConceptCloud:
                        ConceptCloud[b]=1
                    else:
                        ConceptCloud[b]+=1

    for e in ConceptCloud:
        b2 += ConceptCloud[e]#　計算文字雲的字彙量

    if b1==b2:
        with open(ConceptCloudFile,'w') as f:#　寫進Json檔裡
            f.write(json.dumps(ConceptCloud))
        print 'Attr :',len(CombineConcept),'TotalConcepts :',b2
        print 'Trans To ConceptCloud Dict Successfully!'
        

###########################################################################
#[5]　攤平景點標籤

       
def FlatConcept(CombineConcept,AttrFlatTagFile):
    import json
    c1=0
    c2=0
    FlatC={}
    for i in CombineConcept:
        FlatC[i]=[]
        for a in CombineConcept[i]:
            for b in CombineConcept[i][a]:
                c1+=1
                FlatC[i].append(b)
                
    for i in FlatC:
        c2 += len(FlatC[i])
        
    if len(CombineConcept)==len(FlatC) and c1==c2:
        print 'Flat Successfully!'
        JF = json.dumps(FlatC)#　丟進一JSON格式的變數
        with open(AttrFlatTagFile,'w') as f:#　寫進Json檔裡
            f.write(JF)
```
